```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from statsmodels.formula.api import ols
```

```python
df_house = pd.read_csv('kc_house_data.csv')
```

```python
df_house
```

```python
df_house.columns
```

```python
# Define the problem
outcome = 'Price'

x_cols = ['id', 'date', 'bedrooms', 'bathrooms', 'sqft_living',
          'sqft_lot', 'floors', 'waterfront', 'view', 'condition',
          'grade', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated',
          'zipcode','lat', 'long', 'sqft_living15', 'sqft_lot15']
```

```python
predictors = '+'.join(x_cols)
predictors
```

```python
# Fitting the actual model
formula = outcome + '~' + predictors
```

```python
model = ols(formula=formula, data=df_house).fit()
model.summary()
```

```python
# import df with outliers from price and bedrooms removed 
house_clean = pd.read_csv('housing_outliers_removed.csv')
```

```python
#Checking for Colinarity 
co_lin = house_clean.corr().abs().stack().reset_index().sort_values(0, ascending=False)

co_lin['pairs'] = list(zip(co_lin.level_0, co_lin.level_1))

co_lin.set_index(['pairs'], inplace = True)

co_lin.drop(columns=['level_1', 'level_0'], inplace = True)

# cc for correlation coefficient
co_lin.columns = ['cc']

co_lin.drop_duplicates(inplace=True)

co_lin[(co_lin.cc>.7) & (co_lin.cc<1)]
```
