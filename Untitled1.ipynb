```python
# import df with outliers from price and bedrooms removed 
df = pd.read_csv('housing_outliers_removed.csv')
```

```python
continuous = ['id', 'date', 'price', 'bedrooms', 'bathrooms',
              'sqft_living', 'sqft_lot', 'floors', 'sqft_above', 'sqft_basement',
              'yr_built', 'yr_renovated','zipcode', 'lat', 'long', 'sqft_living15',
              'sqft_lot15']

categoricals = ['grade', 'condition', 'waterfront', 'view']

df_cont = df[continuous]
```

```python
# one hot encode categoricals
df_ohe = pd.get_dummies(df[categoricals], prefix=categoricals, drop_first=True)
```

```python
regression = LinearRegression()

preprocessed = pd.concat([df_cont, df_ohe], axis=1)

X = preprocessed.drop('price', axis=1)
y = preprocessed['price']

crossvalidation = KFold(n_splits=10, shuffle=True, random_state=1)
baseline = np.mean(cross_val_score(regression, X, y, scoring='r2', cv=crossvalidation))

baseline
```
